С датасетом с игроками FIFA проделать следующие действия:
Найти количество игроков заданного ФК моложе заданного возраста. Предусмотреть пользовательский ввод клуба и цифры возраста для поиска
По заданному ФК найти средний вес, рост и возраст игроков
По трём любым клубам составить убывающий список игроков с указанием даты присоединения к клубу. Учитывать не только количество лет, а также и дни в году. Т.е. если один игрок присоединился к клубу 15 июня 2020, а второй 15 августа 2020, то первый должен появиться в списке раньше.
Построить диаграмму рассеяния скорость игрока / ускорение игрока (поля movement sprint speed и movement acceleration) по трём любым странам. Если график слишком плотный, уменьшите число игроков. Сделать выводы по графику
Построить 3 ящика с усами по качеству паса игроков любой страны. Отдельно построить для игроков до 24 лет, 24 - 30 лет, 30+ лет. Сделать выводы по графикам
Построить графики зависимости заработной платы игрока от его трансферной стоимости. По трём клубам: топ-клуб, средний в таблице и клуб, близкий к вылету из лиги. Сделать выводы
По трём странам построить графики зависимости заработной платы от возраста игрока. Сделать выводы.
